```python
# import chromadb and create a client
import chromadb

client = chromadb.Client()
collection = client.create_collection("my-collection")
```

```python
# add the documents in the db
collection.add(
    documents=["This is a document on LLM capabilities", "This is a document about overview of llm","This is a document about BERT"],
    metadatas=[{"category": "Research"}, {"category": "Explanation"},{"category": "BERT"}],
    ids=["id1", "id2","id3"]
)
```

```python
# ask the querying to retrieve the data from DB
results = collection.query(
    query_texts=["limitations"],
    n_results=1
)
results
```

```python
# import files from the collections folder to store in VectorDB
import os

def read_files_from_folder(folder_path):
    file_data = []

    for file_name in os.listdir(folder_path):
        if file_name.endswith(".txt"):
            with open(os.path.join(folder_path, file_name), 'r') as file:
                content = file.read()
                file_data.append({"file_name": file_name, "content": content})

    return file_data

folder_path = "/content/drive/MyDrive/LLMFILES"
file_data = read_files_from_folder(folder_path)
```

```python
file_data
```

```python
# get the data from file_data and create chromadb collection
documents = []
metadatas = []
ids = []

for index, data in enumerate(file_data):
    documents.append(data['content'])
    metadatas.append({'source': data['file_name']})
    ids.append(str(index + 1))

# create a collection of llm files
llm_collection = client.create_collection("llm_collection")

# Add files to the chromadb collection
llm_collection.add(
    documents=documents,
    metadatas=metadatas,
    ids=ids
)
```

```python
# query the database to get the answer from vectorized data
results = llm_collection.query(
    query_texts=["What is the limitation of llm"],
    n_results=1
)

results
```

```python
# import the sentence transformers
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('paraphrase-MiniLM-L3-v2')

documents = []
embeddings = []
metadatas = []
ids = []

# enumerate through file_data to collection each document and metadata
for index, data in enumerate(file_data):
    documents.append(data['content'])
    embedding = model.encode(data['content']).tolist()
    embeddings.append(embedding)
    metadatas.append({'source': data['file_name']})
    ids.append(str(index + 1))

# create the new chromaDB and use embeddings to add and query data
llm_collection_emb = client.create_collection("llm_collection_emb")

# add the pets files into the pet_collection_emb database
llm_collection_emb.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=ids
)
```

```python
# write text query and submit to the collection
query = "What are the different kinds of llm's?"
input_em = model.encode(query).tolist()

results = llm_collection_emb.query(
    query_embeddings=[input_em],
    n_results=1
)
results
```
